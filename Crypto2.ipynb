```python
for elem in text:
  if elem not in alphabet:
    text = text.replace(elem, '')
print(text)
```

```python
monogram_frequency = {}
for elem in alphabet:
  monogram_frequency[elem] = 0
for elem in text:
  monogram_frequency[elem]+=1
print(monogram_frequency)
```

```python
def Vigenere_cypher(L):
  import random
  key = []
  key_length = int(input('Введите длину ключа : '))
  for i in range(key_length):
    symbol = random.randint(0,31)
    key.append(alphabet[symbol])
  print(f'key = {key}')
  if len(L)==10000:
    cypher = [0]*10000
    for i in range(len(cypher)):
      cypher[i] = [0]*len(L[0])
    for i in range(len(L)):
      for j in range(len(L[i])):
        num1 = alphabet.index(L[i][j])
        num2 = alphabet.index(key[j%len(key)])
        result = (num1 + num2)%32
        symbol = alphabet[result]
        cypher[i][j] = symbol
  else:
    cypher = [0]*1000
    for i in range(len(cypher)):
      cypher[i] = [0]*len(L[0])
    for i in range(len(L)):
      for j in range(len(L[i])):
        num1 = alphabet.index(L[i][j])
        num2 = alphabet.index(key[j%len(key)])
        result = (num1 + num2)%32
        symbol = alphabet[result]
        cypher[i][j] = symbol
  return cypher
C = Vigenere_cypher(L10000)
print(f'Открытый текст = {L10000[0]}')
print(f'Зашифрованный текст = {C[0]}')


```

```python
bigram_alphabet = []
for i in range(len(alphabet)):
  for j in range(len(alphabet)):
    string = alphabet[i] + alphabet[j]
    bigram_alphabet.append(string)

def affine_bigram(L):
  import random
  a = int(input('Введите значение a : '))
  b = int(input('Введите значение b : '))
  print(f'a = {a}, b = {b}')
  if len(L)==10000:
    cypher = [0]*10000
    for i in range(len(cypher)):
      cypher[i] = [0]*(len(L[i])//2)
    for i in range(len(L)):
      for j in range(0, len(L[i]) - 1, 2):
        for k in range(j+1, j+2):
          symbol = L[i][j] + L[i][k]
          X = bigram_alphabet.index(symbol)
          Y = a*X + b
          Y = Y%1024
          result = bigram_alphabet[Y]
          cypher[i][j//2] = result
  else:
    cypher = [0]*1000
    for i in range(len(cypher)):
      cypher[i] = [0]*(len(L[i])//2)
    for i in range(len(L)):
      for j in range(0, len(L[i]) - 1, 2):
        for k in range(j+1, j+2):
          symbol = L[i][j] + L[i][k]
          X = bigram_alphabet.index(symbol)
          Y = a*X + b
          Y = Y%1024
          result = bigram_alphabet[Y]
          cypher[i][j//2] = result
  return cypher

C = affine_bigram(L10000)
print(f'Открытый текст : {L10000[0]}')
print(f'Зашифрованный текст : {C[0]}')

```

Сформируем множество A_frq для часто встречающихся монограмм

```python
print(monogram_frequency)
def function(dictionary):
  temp = {}
  elements = []
  for elem in dictionary:
    elements.append(dictionary[elem])
  elements.sort()
  elements.reverse()
  for elem in dictionary:
    if dictionary[elem] >= elements[len(elements)//2]:
      temp[elem] = dictionary[elem]
  return temp
A_frq = function(monogram_frequency)
```
