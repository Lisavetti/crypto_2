```python
file = open('Снафф.txt', encoding = 'utf-8')
alphabet = 'абвгдеєжзиіїйклмнопрстуфхцчшщьюя'
text = ''
for row in file:
  text+=row
text = text.replace('\n','')
text = text.replace('ґ','г')
text = text.lower()
print(text)
```

```python
for elem in text:
  if elem not in alphabet:
    text = text.replace(elem, '')
print(text)
```

```python
monogram_frequency = {}
for elem in alphabet:
  monogram_frequency[elem] = 0
for elem in text:
  monogram_frequency[elem]+=1
print(monogram_frequency)
```

```python
sum = 0
for elem in monogram_frequency:
  sum+=monogram_frequency[elem]
```

```python
monogram_distribution = {}
for elem in monogram_frequency:
  monogram_distribution[elem] = monogram_frequency[elem]/sum
print(monogram_distribution)
```

```python
bigram_frequency = {}
for first in alphabet:
  for second in alphabet:
    letter = first + second
    bigram_frequency[letter] = 0
```

```python
for i in range(len(text)-1):
  symbol = text[i] + text[i+1]
  bigram_frequency[symbol]+=1
sum = 0
for elem in bigram_frequency:
  sum+=bigram_frequency[elem]
bigram_distribution = {}
for elem in bigram_frequency:
  bigram_distribution[elem] = bigram_frequency[elem]/sum
#print(bigram_distribution)
```

```python
import math
monogram_enthropy = 0
for elem in monogram_distribution:
  monogram_enthropy+=monogram_distribution[elem]*math.log2(monogram_distribution[elem])
monogram_enthropy = -1 * monogram_enthropy
monogram_enthropy = 1/len(monogram_frequency) * monogram_enthropy
print(monogram_enthropy)
```

```python
import math
bigram_enthropy = 0
#print(bigram_distribution)
b_dist_new = {}
for elem in bigram_distribution:
  if bigram_distribution[elem]!=0:
    b_dist_new[elem] = bigram_distribution[elem]
for elem in b_dist_new:
  bigram_enthropy+=b_dist_new[elem]*math.log2(b_dist_new[elem])
bigram_enthropy = (-1)*bigram_enthropy
bigram_enthropy = 1/len(bigram_distribution) * bigram_enthropy
print(bigram_enthropy)
```

```python
Index_Mono = 0
print(monogram_frequency)
for elem in monogram_frequency:
  Index_Mono+=monogram_frequency[elem] * (monogram_frequency[elem] - 1)
Index_Mono = 1/(len(monogram_frequency) * (len(monogram_frequency) - 1)) * Index_Mono
print(Index_Mono)
```

Реализация афинной биграммной подстановки

```python
bigram_alphabet = []
for i in range(len(alphabet)):
  for j in range(len(alphabet)):
    string = alphabet[i] + alphabet[j]
    bigram_alphabet.append(string)


a = int(input('Введите значение a : '))
b = int(input('Введите значение b : '))

cypher_L10 = [0]*10000
for i in range(len(cypher_L10)):
  cypher_L10[i] = [0]*5

for i in range(len(L10)):
  for j in range(0, len(L10[i]) - 1, 2):
    for k in range(j+1,j+2):
      symbol = L10[i][j] + L10[i][k]
      X = bigram_alphabet.index(symbol)
      Y = a*X + b
      Y = Y%1024
      result = bigram_alphabet[Y]
      cypher_L10[i][j//2] = result

print(L10[0])
print(cypher_L10[0])
```
