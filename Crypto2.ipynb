```python
import zlib
import random

def structure_criteria(text, limit = 0.1):
  H1 = 0
  text_len = len(text)
  length = len(text[0])
  massive = [''.join([alphabet[random.randint(0,31)] for i in range(length)]) for i in range(text_len)]
  random_values = []
  for elem in massive:
    a = length / len(zlib.compress(elem.encode('utf-8')))
    random_values.append(a)
  real_values = []
  for elem in text:
    a = length / len(zlib.compress(elem.encode('utf-8')))
    real_values.append(a)
  for i in range(len(random_values)):
    if abs(random_values[i] - real_values[i]) > limit:
      H1 = H1 + 1
  return f'H1 = {H1}, H0 = {len(random_values) - H1}'

print(structure_criteria(L100))
print(structure_criteria(Vigenere_bi(L100)))
```
