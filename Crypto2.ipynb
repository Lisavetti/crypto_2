```python
print(f'Частота : {bigram_frequency}')
print(f'B_frq = {B_frq}')

def bigram_decompresion(L):
  massive = [0]*len(L)
  for i in range(len(massive)):
    massive[i] = [0]*(len(L[i])//2)
  for i in range(len(L)):
    for j in range(0, len(L[i]) - 1, 2):
      massive[i][j//2] = L[i][j] + L[i][j+1]
  return massive


def criteria_20_bigram(L, B_frq):
  L = bigram_decompresion(L)
  H0 = 0
  H1 = 0
  massive = []
  for key in B_frq:
    massive.append(key)
  for i in range(len(L)):
    for elem in massive:
      if elem not in L[i]:
        H1 = H1 + 1
        break
  return f'H1 = {H1}'

print(len(B_frq))
print(criteria_20_bigram(L1000, B_frq))
```
